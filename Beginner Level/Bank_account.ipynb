```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Date": [
        "2025-01-05", "2025-01-10", "2025-01-15", "2025-01-18", "2025-01-22",
        "2025-01-25", "2025-02-01", "2025-02-07", "2025-02-10", "2025-02-14",
        "2025-02-20", "2025-02-23", "2025-02-27", "2025-03-02", "2025-03-05"
    ],
    "Type": [
        "Deposit", "Withdraw", "Deposit", "Withdraw", "Deposit",
        "Withdraw", "Deposit", "Withdraw", "Deposit", "Withdraw",
        "Deposit", "Withdraw", "Deposit", "Withdraw", "Deposit"
    ],
    "Amount": [
        1500.00, 200.00, 1200.00, 100.00, 900.00,
        400.00, 2500.00, 300.00, 1800.00, 150.00,
        1000.00, 350.00, 2200.00, 500.00, 1700.00
    ]
}
df = pd.DataFrame(data, index = data["Date"])
df
```

```python
largest_transaction = df.groupby("Type")["Amount"].max()
print(largest_transaction)
```

```python
total_transaction_type = df.groupby("Type").size()
print(total_transaction_type)
```

```python
df ["Signed Amount"] = df.apply(
                    lambda row: row["Amount"] if row["Type"] == "Deposit" else -row["Amount"],
                    axis = 1)
df["Balance"] = df["Signed Amount"].cumsum()
print(df[["Type", "Amount", "Balance"]])
```

```python
low_balance = df[df["Balance"] < 4000]["Balance"]
print(low_balance)
```
