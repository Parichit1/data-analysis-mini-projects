```python
import numpy as np
import pandas as pd
```

```python
students = ["Pam", "Jim", "Michael", "Dwight", "Toby"]
courses = ["Math", "English", "Science"]
```

```python
marks = np.random.randint(40,101, size = (len(students), len(courses)))
print(marks)
```

```python
avg_studnet = np.mean(marks, axis=1)
print(avg_studnet)
```

```python
avg_course = np.mean(marks, axis=0)
print(avg_course)
```

```python
sd_studnet = np.std(marks, axis=1)
print(sd_studnet)
```

```python
sum_grades = np.sum(marks, axis=1)
max_grades = np.max(sum_grades)
print(max_grades)
```

```python
df = pd.DataFrame(marks, index= students, columns= courses)
print(df)
```

```python
avg_grades_persub = df.median(axis=0)
print(avg_grades_persub)
```

```python
totalscore = df.sum(axis=1)
student_highestscore = totalscore.idxmax() 
print(student_highestscore)
```

```python
fail_student = df<50
failed_any = fail_student.any(axis=1)
total_failed_any = failed_any.sum()
print(total_failed_any)
```
