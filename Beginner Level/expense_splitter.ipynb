```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Date": pd.date_range(start="2025-10-01", periods=25, freq="D"),
    "Category": [
        "Groceries", "Rent", "Transport", "Entertainment", "Utilities",
        "Dining Out", "Groceries", "Rent", "Transport", "Healthcare",
        "Entertainment", "Groceries", "Rent", "Dining Out", "Subscriptions",
        "Transport", "Groceries", "Utilities", "Entertainment", "Rent",
        "Dining Out", "Transport", "Groceries", "Healthcare", "Subscriptions"
    ],
    "Amount": [
        120.45, 1500.00, 45.60, 75.20, 200.00,
        60.75, 130.10, 1500.00, 42.50, 90.00,
        100.00, 125.30, 1500.00, 55.80, 12.99,
        38.40, 110.25, 210.00, 80.50, 1500.00,
        62.00, 40.10, 95.25, 85.75, 14.99
    ],
    "Paid_By": [
        "Patrick", "Alex", "Jordan", "Patrick", "Casey",
        "Alex", "Jordan", "Patrick", "Casey", "Alex",
        "Jordan", "Patrick", "Casey", "Alex", "Jordan",
        "Patrick", "Casey", "Alex", "Jordan", "Patrick",
        "Casey", "Alex", "Jordan", "Patrick", "Casey"
    ]
}
df = pd.DataFrame(data)
print(df)
```

```python
spent_by_category = df.groupby("Category")["Amount"].sum()
print(spent_by_category)
```

```python
spent_by_person = df.groupby("Paid_By")["Amount"].sum()
print(spent_by_person)
```

```python
avg_spent_by_person = spent_by_person.mean()
print(avg_spent_by_person)
```

```python
balance = spent_by_person - avg_spent_by_person
print(round(balance, 2))
```
