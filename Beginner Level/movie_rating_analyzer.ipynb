```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Title": [
        "Inception", "The Dark Knight", "Interstellar", "Parasite", "The Godfather",
        "Pulp Fiction", "The Shawshank Redemption", "Whiplash", "The Matrix", "Gladiator",
        "La La Land", "Joker", "The Social Network", "Mad Max: Fury Road", "The Grand Budapest Hotel",
        "Get Out", "Dune", "Spider-Man: Into the Spider-Verse", "The Irishman", "Everything Everywhere All at Once"
    ],
    "Genre": [
        "Sci-Fi", "Action", "Sci-Fi", "Thriller", "Crime",
        "Crime", "Drama", "Drama", "Sci-Fi", "Action",
        "Musical", "Drama", "Biography", "Action", "Comedy",
        "Horror", "Sci-Fi", "Animation", "Crime", "Sci-Fi"
    ],
    "IMDb_Rating": [
        8.8, 9.0, 8.6, 8.6, 9.2,
        8.9, 9.3, 8.5, 8.7, 8.5,
        8.0, 8.4, 7.8, 8.1, 8.1,
        7.8, 8.0, 8.4, 7.9, 8.0
    ],
    "Release_Year": [
        2010, 2008, 2014, 2019, 1972,
        1994, 1994, 2014, 1999, 2000,
        2016, 2019, 2010, 2015, 2014,
        2017, 2021, 2018, 2019, 2022
    ],
    "Duration_Minutes": [
        148, 152, 169, 132, 175,
        154, 142, 107, 136, 155,
        128, 122, 120, 120, 99,
        104, 155, 117, 209, 139
    ]
}
df = pd.DataFrame(data)
df
```

```python
average_rating_genre = round(df.groupby("Genre")["IMDb_Rating"].mean(), 2)
print(average_rating_genre)
```

```python
avg_duration = df["Duration_Minutes"].mean()
print(avg_duration)
```

```python
movies_longer_avg_durartion = df[df["Duration_Minutes"] > avg_duration] [["Title", "Duration_Minutes"]]
print(movies_longer_avg_durartion)
```

```python
idx = df.groupby("Genre")["IMDb_Rating"].idxmax()
highest_rated_in_genre = df.loc[idx, ["Genre", "Title", "IMDb_Rating"]]
print(highest_rated_in_genre)
```

```python
conditions = [
    df["IMDb_Rating"] > 8,
    ((df["IMDb_Rating"] > 6) & (df["IMDb_Rating"] <= 8)),
    df["IMDb_Rating"] <= 6
]

choices = ["Excellent", "Good", "Average"]

df["Rating_Category"] = np.select(conditions, choices, default="Unknown")
print(df[["Title", "IMDb_Rating", "Rating_Category"]])
```
