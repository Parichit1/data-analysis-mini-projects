```python
import pandas as pd
```

```python
data = {"Date" : ["1st feb", "2nd feb", "3rd feb", "4th feb"],
                "Account": ["Savings","Chequing", "Chequing","Savings"],
                "Transaction_type": ["Deposit", "Withdrawal","Deposit", "Withdrawal"],
                "Amount": [500, -150, 300, -100]}

df = pd.DataFrame(data)
print(df)
```

```python
df.loc[4] = {"Date": "5th feb",
                    "Account": "Chequing",
                    "Transaction_type": "Deposit",
                    "Amount": 250}
print(df)
```

```python
total_amount = df["Amount"].sum()
print(total_amount)
```

```python
amount_bygroup = df.groupby("Account")["Amount"].sum()
print(amount_bygroup)
```

```python
max_transaction = df["Amount"].max()
print(max_transaction)
```
