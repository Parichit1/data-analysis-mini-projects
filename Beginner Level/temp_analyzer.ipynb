```python
import numpy as np
import pandas as pd
import random
```

```python
temp = np.random.uniform(0.00, 40.00, size = 7)
temp = np.round_(temp, 2)
print(temp)
```

```python
temp_far = (temp *(9/5) + 32)
temp_far = np.round_(temp_far,2)
print(temp_far)
```

```python
temp_weekavg = np.mean(temp)
temp_weekavg = np.round_(temp_weekavg,2)
temp_weekavg_far = np.mean(temp_far)
temp_weekavg_far = np.round_(temp_weekavg_far,2)
print(f"Weekly avegrage in Cel:  {temp_weekavg}")
print(f"Weekly avegrage in Far:  {temp_weekavg_far}")
```

```python
temp_min_cel = np.min(temp)
print(f"Min temp in Cel: {temp_min_cel}")
temp_min_far = np.min(temp_far)
print(f"Min temp in Far: {temp_min_far}")
```

```python
temp_max_cel = np.max(temp)
print(f"Max temp in Cel: {temp_max_cel}")
temp_max_far = np.max(temp_far)
print(f"Max temp in Far: {temp_max_far}")
```

```python
above_avg_cel = temp[temp > temp_weekavg]
count_above_avg_cel = len(above_avg_cel)
print(f"So there are total of {count_above_avg_cel} days above the weekly avg and those are {above_avg_cel}")
```

```python
above_avg_far = temp_far[temp_far>temp_weekavg_far]
count_above_avg_far = len(above_avg_far)
print(f"So there are total of {count_above_avg_far} days above the weekly avg and those are {above_avg_far}")
```
